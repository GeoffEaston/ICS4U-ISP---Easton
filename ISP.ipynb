<h1>Read in the cvs dataset</h1>
<hr>
This data set includes secondary school acceptance averages, graduation rates, retetion rates, and headcounts over the 4 years of an unddergraduate program, as well as total applicants and registrants. This data was obtained from the school's respective Common University Data Ontario (CUDO) reports, spanning from 2015 to 2019, via the Ontario Universities website. The data involving the headcounts of students over the 4 years of an undergraduate program was obtained from the Ontario government website's data on university enrolment. The schools recorded in this dataset include the University of Toronto, Brock, Western, Waterloo, Guelph, and Carleton.

It's worth noting that some values and even entire programs were sometimes not stated by universities in their CUDO reports, such as for acceptance averages and graduation rates. This was usually done for privacy and other reasons. Thus, unknown values such as these are denoted as hypens on the data set and are treated differently depending on the figure. For example, in the bar graphs representing the distribution of acceptance averages, these unknown values are treated as 0.00% and are not shown. And for the box chart representing graduation rates, these values were removed entirely from the data used.

```python
# Load the data
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Comp Sci ISP Data Sets - Sheet1.csv")
data.head()
```

<h1>Acceptance Averages - Bar graphs</h1>

```python
def toInt(l):
    for i in range(len(l)):
        l[i] = l[i][:-1]
        l[i] = float(l[i])
    return(l)

def toList(s):
    for i in range(s.index[0],s.index[0]+7,1):
        if s[i]=='-':
            s[i]= '0.00%'
    l = []
    for j in range(s.index[0],s.index[0]+7,1):
        l.append(s[j])
    l.reverse()
    toInt(l)
    print(l)
    return l

def barPlot(s):
    l = toList(s)
    c = ('>70%','70%-74%','75%-79%','80%-84%','85%-89%','90%-94%','95%+')
    plt.bar(c,l)
    plt.ylim(0.00,100.00)
    t = 'Secondary school averages of incoming first year population, '
    plt.title(t+s.name)
    plt.ylabel('Percentage of population')
    plt.show()
 
barPlot(data.iloc[109:116,2])
barPlot(data.iloc[109:116,3])
barPlot(data.iloc[109:116,4])
barPlot(data.iloc[109:116,5])
barPlot(data.iloc[109:116,6])
barPlot(data.iloc[109:116,7])
```

The figures above shows the distribution of the secondary school averages of an incoming class of first year undergraduate students for each of the 6 universities. In short, it shows what percentage of accepted first year students had a certain grade average in high school. This goes into further detail with the acceptance averages of specific programs. The programs included in this portion of the data set include Architecture, Arts, the Sciences, Business & Commerce, Physical Education, Engineering, Fine Arts, Music, Mathematics, Nursing, Journalisms, and other courses, as well as the school overall. 

The code above has three primary operations: to read the data set, to clean up the specified data, and to build the bar charts out of it. First, the barPlot function is called with a specific series from the data set as its parameters. This function then creates a list from this series using the toList function, which replaces any '-' with 0.00% and converts all elements from stings to floats. Finally, a bar graph is created using the list of float values from the data. 

Bar graphs are being used in this case specifically because the bar's heights can better represent the demographic of first year students a university accepts based on their high school marks. This usually results in a standard distribution which can show which averages a university accepts the most of as well as whether or not they skew towards higher or lower marks. A histogram, for example, wasn't used because they display the frequency of a data set while a bar graph compares different categories. If the secondary school averages of every student were available, a histogram could be used and would look similar to these bar graphs, but the proportion of the first year population within a certain range of marks were available instead so a bar graph is used. This all can be very helpful to university-bound high school students to show what averages they need to aim for in order to have a better chance of getting into their university and or program of choice.

The bar graphs above were created using the distribution of high school averages for each university overall. This was done to limit the amount of unknown values and visualize the trends that the individual program's will likely be similar to. What we see here is that the majority of students that are accepted to schools like Brock, Guelph, and Carleton are those with averages of 80% to 84%. Their accepted demographic also forms a standard distribution with students with higher and lower averages making up less, but equal parts of the population. These distributions show us that these schools are easier to get into so long as you have an average around 80%. Also, since students with higher than average marks (>85%) make up a smaller part of the population, students with these kinds of marks choose to go elsewhere for their post-secondary studies - these schools may not be the best for more competitive programs.The University of Toronto is similar, but skews slightly towards higher marks by accepting more students with 85%-89%. Waterloo and Western skew the furthest to the right, showing that these schools are a lot tougher to get into without an average above 89% but offer very high quality for the more competitive programs. To further stress how difficult it is to get into these schools, Western rarely accepts those with averages below 80%. Those below 80% make up only 0.2% of incoming undergraduate students to Western.

<h1>Graduation Rates - Box plots</h1>

```python
def toFloat(s):
    for i in s.index.tolist():
        s[i]= s[i][:-1]
        s[i]=float(s[i])
    return s

def boxPlot(school):
    s = data.loc[data.iloc[:,10]==school]
    s1= s['Graduation Rate']
    l=[]
    for i in s1.index.tolist():
        if s1[i]=='-':
            l.append(i)
    s2 = s1.drop(l)
    s2 = toFloat(s2)
    return s2

l = [boxPlot('UoT'),boxPlot('Brock'),boxPlot('Western'),boxPlot('Waterloo'),boxPlot('Guelph'),boxPlot('Carleton')]

plt.title('Graduation Rates')
plt.boxplot(l)
plt.xticks([1,2,3,4,5,6],['Uot','Brock','Western','Waterloo','Guelph','Carleton'])
plt.ylabel('Percentage')
plt.grid(True)
plt.show()
```

The box plots above represent the range of graduation rates for the programs at each respective school. These programs include Architecture, Business & Commerce, Computer Science, Engineering, Fine & Applied Arts, Humanities, Physical Education, Mathematics, Nursing, and other Arts & Sciences as well as the completion rates for the schools overall.

The code above uses the boxPlot function to clean up Series of data from the data set by removing any unknown values and converting the rest into floats that can be plotted on the box plot.

A boxplot was used to present the school's completion rates because box charts can better represent the different ranges of difficulty each school offers. A box plot presents a range of data's maximum and minimum values, as well as its median and first and third quartiles (the average of the first and second half of the data). With these values, box plots can help us better understand the composition of a range of data. The higher a box plot is on the y-axis, similar to headcount over time, the better chances a student has at successfully getting a degree from a program from a specific university.Also, the dots seen on the graph above are outliers which do not fit into the range created by the boxplot. Overall, what indicates a school where a student can be more successful is a box plot that is higher along the y-axis - its average and minimum rates are higher with a thinner box (programs are more similar to the average's rate of completion).


The University of Toronto has the highest graduation rates among these six universities - at 98.2% for their Nursing program and 100% for Humanities which is their highest outlier - but its average and minimums (62.5% in Mathematics, the second outlier, and 73.9% in other Arts & Sciences, the minimum) are not the highest themselves. This shows that although individually some of their programs may be the most successful in a certain field, overall students will have a harder time completion (see the larger distance between UoT's maximum and its 3rd quartile). The school with the highest average completion rate is Western, only slightly higher than UoT's ~85%. It's boxplot is also very thin - although its maximum completion rate is lower than UoT's,but it has a minimum rate ~80% - with all of the recorded programs falling within 80%-90% with the average closer to the maximum. This shows that Western is a safer choice overall to complete your degree. The university that shows the poorest range of graduation rates on this box plot would be Carleton. Although it has the second highest maximum (92.9%, Architecture), it also has the widest range of graduation rates. It's minimum rate is 48.7% for Computer science and has the longest distance between its maximum rate and 3rd quartile, which shows that the high completion rate of its maximum is unique only to its Architecture program and that its programs overall are closer to its average of about 68%, which is the lowest among the six schools.

<h1>Part 2: Data Analysis - Decision Trees</h1>

<hr>

```python
data2 = pd.read_csv("ISP Data set 2.csv")
data2.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text


Programd = {'Architecture':1, 'Arts':2, 'Science':3, 'Business & Commerce':4, 'Physical Education':5, 'Engineering':6, 'Fine Arts':7, 'Music':8, 'Mathematics':9, 'Nursing':10, 'Journalism':11, 'Other':12, 'Overall':13}
Schoold = {'UoT':1, 'Brock':2,'Western':3, 'Waterloo':4, 'Guelph':5, 'Carleton':6}

def fix(s):
    for i in range(0,78):
        for j in range(1,8):
            if s.iat[i,j]!='0':
                s.iat[i,j]= s.iat[i,j][:-1]
    for k in range(0,78):
        s.iat[k,0]=Programd[s.iat[k,0]]
    s = s.astype(float)
    return s

def Schooltofloat(s):
    for i in range(0,78):
        s.iat[i,0]=Schoold[s.iat[i,0]]
    s = s.astype(float)
    return s


X = data2.iloc[:,[0]] ##X - Schools
Y = data2.iloc[:,[1,2,3,4,5,6,7,8]] ##Y - Programs & Average's range
Y = fix(Y)
print(X)
print(Y)

X_train, X_test, Y_train, Y_test = train_test_split(Y, X, test_size=0.3)
model = DecisionTreeClassifier(max_depth=4)
clf = model.fit(X_train, Y_train)

plt = tree.plot_tree(clf,feature_names=['Program', '95%+','90%-94%','85%-89%','80%-84%','75%-79%','70%-74%','>70%'],  class_names=["UoT","Brock","Western","Waterloo","Guelph","Carleton"],filled=True)

r = export_text(clf)
print(r)
```

The code above, although confusing to all hell, especially at the end, takes the data from the second ISP data set (which is just a reformated version of table 1 from the first data set) and cleans it up before using it to make a Decision Tree out of it. The goal of the decision tree is to predict which school a student would have the best chance of getting into based on their choice of program and high school average. Essentially by specifying a student's choice of program and average it checks the data for which of the 6 schools has the largest incoming population with that range of averages.

The 'fix' and 'Schooltofloat' functions are used to clean up the data that's used for the Decision tree. They remove the percentage signs from the data, as well as number the Programs(turning them into numerical data that's useable by the tree) and converting it all from strings to floats. From there, the dataframe and series of schools are used to train the Decision tree. For example, if a student has an 83% average and wants to go into Engineering, the decision tree will compare their mark with the range of marks specified in the data (as well as identify which program 'engineering' is) until its found the school who has the largest population of engineering students with marks within the same range as them. If said school has the most engineering students within that range of marks then the student has the best chance of getting into it out of all the other schools. This school would actually end up being Carleton, where 30.70% of incoming undergraduate students have a high school average with the range of 80% to 84%.
